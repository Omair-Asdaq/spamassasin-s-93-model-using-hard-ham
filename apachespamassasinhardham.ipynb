```python
#achieves 93.3 % on test (could be better than now)
#used 

#https://colab.research.google.com/github/PseudoCodeNerd/blog/blob/master/_notebooks/2019-10-19-spamClassifier-Oreilly-homework-chapter3.ipynb#scrollTo=0Sg4Mbnj-7yA

#for the get FUNCTION only ()

import tarfile
import os
import urllib.request

down_path = "http://spamassassin.apache.org/old/publiccorpus/"
hard_ham_url = down_path + "20030228_hard_ham.tar.bz2"
spam_url = down_path + "20030228_spam.tar.bz2"
spam_path = os.path.join("datasets", "spam")

def fetch_spam_data():
    if not os.path.isdir(spam_path):
        os.makedirs(spam_path)

    files = [
        ("hard_ham.tar.bz2", hard_ham_url),
        ("spam.tar.bz2", spam_url)
    ]

    for filename, url in files:
        path = os.path.join(spam_path, filename)

        if not os.path.isfile(path):
            urllib.request.urlretrieve(url, path)

        with tarfile.open(path) as tar:
            tar.extractall(path=spam_path)

    print("Download and extraction complete.")
```

```python
def cleanandseparate(s):
    import re

    # Extract headers first didnt use these in the pipeline but extracted them if somebody wants to try them 
    Subject_match = re.search(r"(?i)^\s*subject:\s*(.*)$", s, re.MULTILINE)
    Subject = Subject_match.group(1) if Subject_match else ""

    To_match = re.search(r"(?i)^\s*to:\s*(.*)$", s, re.MULTILINE)
    To = To_match.group(1) if To_match else ""

    From_match = re.search(r"(?i)^\s*from:\s*(.*)$", s, re.MULTILINE)
    From_ = From_match.group(1) if From_match else ""

    # Lowercase and clean removing all punctuations etc 
    s = s.lower()
    s = re.sub(r'[()\[\]{};:"]+', "", s)
    s = re.sub(r"\d+", "NUMBER", s)
 
    # Replace URLs had gpt write this
    url_pattern = re.compile(
        r"\b(?:https?://|http//|ftp://|www\.)"
        r"[-a-zA-Z0-9@:%._\+~#=]{1,256}"
        r"\.[a-zA-Z]{2,6}"
        r"(?:[-a-zA-Z0-9()@:%_\+.~#?&//=]*)"
    )
    s = url_pattern.sub("URL", s)

    # Extract body safely
    parts = s.split('\n\n', 1)
    Body = parts[1].strip() if len(parts) > 1 else "" 

    return s, Subject, To, From_, Body
```

```python
# removes all single letters  then tokenizers breaks the string from whitespaces you can remove all the single letters after you split the list
def tokenize(s): 
    s = re.sub(r"\b[a-z<>A-Z]\b", "", s)
    s = re.split(r'[\n ,._!"\'\t-\\]+' , s) 
    s = [l for l in s if l!= '']
    return list(dict.fromkeys(s))
    
```

```python
X = spam_text + ham_text
y = [1]*len(spam_text) + [0]*len(ham_text)  # 1 = spam, 0 = ham
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
#full tokenize takes the full email and converts it into tokens (very messy)
def globaltokenizer(s,fulltokenize = False ):
    
    globallist = []
    tokenlist = []

    for email in s:
        string, Subject, To, From_, Body = cleanandseparate(email)
        tokens = tokenize(email if fulltokenize else Body)          
        tokenlist.append(tokens)
        globallist.extend(tokens)

    unique_tokens = list(dict.fromkeys(globallist))  
    return unique_tokens, tokenlist
```

```python
vocab,toklist = globaltokenizer(X_train)
```

```python
vocab_test,toklist_test = globaltokenizer(X_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# vocab: unique tokens from training set
# toklist: token lists for each training email

vectorizer = CountVectorizer(analyzer=lambda x: x, vocabulary=vocab)
X_train_sparse = vectorizer.transform(toklist)
print("Sparse matrix shape:", X_train_sparse.shape) # check if correctly worked
```

```python
X_test_sparse = vectorizer.transform(toklist_test)
```

```python
from sklearn.naive_bayes import MultinomialNB # testing the naive bayes

clf = MultinomialNB()
clf.fit(X_train_sparse, y_train)
y_pred = clf.predict(X_train_sparse)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

clf = MultinomialNB(alpha = 1,class_prior=[0.000000004, 0.999999996]) #chose these values to test my preprocessing

y_train_pred = clf.predict(X_train_sparse)

print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
print(classification_report(y_train, y_train_pred))
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np
#logisticRegression as chosen model (performs better than naive bayes )
clf = LogisticRegression(
    class_weight="balanced",
    C = 0.01
)
y = y_train
X = X_train_sparse
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(clf, X, y, cv=skf, scoring='f1')
print("Cross-validation f1:", scores)
print("Mean CV f1:", np.mean(scores))
clf.fit(X,y)
y_train_pred = clf.predict(X_train_sparse)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
print(classification_report(y_train, y_train_pred))
```
